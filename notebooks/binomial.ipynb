```python
from numpy.random import binomial
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
admix_prop_auto = 0.088
admix_prop_x = 0.31
```

```python
N = 2 * 15000
startfreq = admix_prop_x
lost = []
fixed = []
replicates = 500
max_generations = 100000
for i in range(replicates):
    p = startfreq
    for g in range(1, max_generations):
        if p == 1:
            fixed.append(g)
            break
        if p == 0:
            lost.append(g)
            break
        p = binomial(N, p, size=1)[0] / N
```

```python
bins = np.arange(0, max_generations, 1000)
plt.hist(lost, bins=bins)
plt.hist(fixed, bins=bins) ;
```

```python
shared_x = np.arange(max(min(lost), min(fixed)), min(max(lost), max(fixed)), 1)

x = np.sort(fixed)
y = np.arange(len(x))/replicates
interp_func = interp1d(x, y)
y_fixed = interp_func(shared_x)

x = np.sort(lost)
y = np.arange(len(x))/replicates
interp_func = interp1d(x, y)
y_lost = interp_func(shared_x)
```

```python
plt.plot(shared_x, y_fixed)
plt.plot(shared_x, y_lost)
plt.axhline(startfreq, linestyle='dashed', color='black')
plt.axhline(0.1, linestyle='dashed', color='black')
plt.axvline(28000, linestyle='dashed', color='black')
plt.ylim(ymin=0)
plt.xlim(xmin=0) ;
```

```python
plt.plot(shared_x, y_fixed/(y_fixed+y_lost))
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.axhline(startfreq, linestyle='dashed', color='black') ;
```
